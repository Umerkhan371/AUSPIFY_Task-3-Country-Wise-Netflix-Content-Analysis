```python
#Import Libraries
```

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('Netflix_dataset.csv')
```

```python
df.head()
```

```python
# Step 1: Extract and clean country information
```

```python
# Check country column
df['country'].head()
```

```python
# Count missing values
df['country'].isnull().sum()
```

```python
# Split multiple countries
df['country'] = df['country'].str.split(',')
```

```python
# Convert list values into separate rows
df = df.explode('country')
```

```python
# Remove extra spaces
df['country'] = df['country'].str.strip()
```

```python
# Verify cleaned data
df['country'].head()
```

```python
# Step 2: Calculate content count by country
```

```python
# Count content for each country
country_count = df.groupby('country').size()

# Sort values in descending order
country_count = country_count.sort_values(ascending=False)

# Display top 10 countries
country_count.head(10)
```

```python
# Step 3: Identify top content-producing countries.
```

```python
top10 = country_count.head(10)

print(top10)
```

```python
# Step 4: Create charts and rankings.
```

```python
# Convert into DataFrame
ranking = top10.reset_index()

ranking.columns = ['Country', 'Content_Count']

# Add Rank Column
ranking['Rank'] = range(1, len(ranking)+1)

print(ranking)
```

```python
plt.figure(figsize=(12,6))

plt.bar(
    top10.index,
    top10.values,
    color='skyblue'
)

plt.title(
    'Top 10 Netflix Content Producing Countries'
)

plt.xlabel('Country')

plt.ylabel('Number of Titles')

plt.xticks(rotation=45)

plt.show()
```

```python
top10.plot(
    kind='barh',
    figsize=(10,6),
    color='orange'
)

plt.title(
    'Top 10 Countries by Netflix Content'
)

plt.xlabel('Content Count')

plt.ylabel('Country')

plt.show()
```

```python
top10.plot(
    kind='pie',
    y='Content_Count',
    autopct='%1.1f%%',
    figsize=(8,8)
)

plt.ylabel('')

plt.title(
    'Netflix Content Share by Top Countries'
)

plt.show()
```
